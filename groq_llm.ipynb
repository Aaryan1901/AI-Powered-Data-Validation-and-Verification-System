```python
import pandas as pd
import time

# Load Excel and prompt
file_path = r"C:\Users\Aaryan\Documents\project-1_materials-20250512T132232Z-1-001\project-1_materials\city_name.xlsx"
sheet_name = 'Cosmo Input Report'
df = pd.read_excel(file_path, sheet_name=sheet_name)

with open(r"C:\Users\Aaryan\Documents\project-1_materials-20250512T132232Z-1-001\project-1_materials\prompt.txt") as f:
    prompt_template = f.read()

def query_with_retry(prompt, retries=5, backoff_factor=2):
    for attempt in range(retries):
        result = query_groq_llm(prompt)
        if not result.startswith("Request failed: 429"):
            return result
        wait_time = backoff_factor ** attempt
        print(f"⚠️ Rate limited. Retrying in {wait_time}s...")
        time.sleep(wait_time)
    return "❌ Failed after retries due to rate limit."

# Process each row
for i, row in df.iterrows():
    main_city = row['Source City name']
    description = row['Description']
    prompt = prompt_template.format(main_city=main_city, description=description)
    
    result = query_with_retry(prompt)
    print(f"[{i}] ✅ Response:\n{result}\n")
```
